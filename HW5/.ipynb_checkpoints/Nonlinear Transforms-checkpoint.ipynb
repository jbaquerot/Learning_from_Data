```python
x1 = np.array([np.random.uniform(-1,1) for i in range(1000)])
x2 = np.array([np.random.uniform(-1,1) for i in range(1000)])
```

```python
plt.plot(x1,x2, 'ro')
```

```python
color = sign( 0.01+ (-0.2*x1**2) + (0.2*x2**2))
color = ['r' if (n>0) else 'b' for n in color]
plt.scatter(x1,x2, color=color)
```
