# Problem 2

## Training the SVC model

```python
def getError( model, X, Y):
    lenght = len(Y)
    #print sum(model.predict(X))
    return 1-(1.0*sum(model.predict(X) == Y)/lenght)
```

```python
def getDataFor(DS, n):
    X_training = DS[['symmetry','intensity']].values
    Y_training = map(lambda x: 1 if (x==n) else -1, DS.digit)
    return X_training, Y_training
```

```python
X, Y = getDataFor(trainingDS, 4)
```

```python
Y
```

```python
def trainingSVM(DS, n):
 
    X_training, Y_training = getDataFor(DS, n)

    polynomial_SVM = svm.SVC(C=0.01, kernel = 'poly', degree = 2, coef0 = 1)

    #training the model
    polynomial_SVM.fit(X_training, Y_training)
    return polynomial_SVM, X_training, Y_training

```

```python
def versusAll(DS, n):
    model, X, Y = trainingSVM(DS, n)
    return getError(model, X, Y)
    
```

```python
print "In sample Error of 0 versus all: ", versusAll(trainingDS, 0)
print "In sample Error of 2 versus all: ", versusAll(trainingDS, 2)
print "In sample Error of 4 versus all: ", versusAll(trainingDS, 4)
print "In sample Error of 6 versus all: ", versusAll(trainingDS, 6)
print "In sample Error of 8 versus all: ", versusAll(trainingDS, 8)
```

# Problem 5

```python
def getDataNvsM(DS, n, m):
    DSOnlyNandM = trainingDS[(trainingDS.digit==n) | (trainingDS.digit==m)]
    X_training = DS[['symmetry','intensity']].values
    Y_training = map(lambda x: 1 if (x==n) else -1, DS.digit)
    return X_training, Y_training
```

```python
X, Y = getDataNvsM(trainingDS, 1, 5)
```

```python
def runExperiment(trainingDS, testDS, n, m, CArg, Q):
    X, Y = getDataNvsM(trainingDS, n, m)
    polynomial_SVM = svm.SVC(C=CArg, kernel = 'poly', degree = Q, coef0 = 1)
    #training the model
    polynomial_SVM.fit(X,Y)
    X_train, Y_train = getDataNvsM(testDS, n, m)
    E_out = getError( polynomial_SVM, X_train, Y_train)
    print "Experient ", n, " vs ", m, "Q =", Q, "C= ", CArg, " #SV", polynomial_SVM.n_support_, "E_out:  ", E_out
```

```python
runExperiment(trainingDS, testingDS, 1, 5, 0.001, 2)
runExperiment(trainingDS, testingDS, 1, 5, 0.01, 2)
runExperiment(trainingDS, testingDS, 1, 5, 0.1, 2)
runExperiment(trainingDS, testingDS, 1, 5, 1, 2)
```
