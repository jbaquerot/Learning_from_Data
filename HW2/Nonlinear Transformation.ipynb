```python
import random
import numpy as np
from numpy.linalg import inv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def createTargetFuntion():
    # f(x1,x2) = sign(x1^2+x2^2-0.6)
    return lambda x1, x2: np.sign(x1**2+x2**2-0.6)


def createTrainingSet(N):
    trainingSet = pd.DataFrame()
    trainingSet['x0'] = np.ones(N)
    trainingSet['x1'] = np.random.uniform(low=-1, high=1, size=N )
    trainingSet['x2'] = np.random.uniform(low=-1, high=1, size=N )
    f = createTargetFuntion()
    trainingSet['f'] = [f(x1,x2) for [x0, x1, x2] in trainingSet.values]
    return f, trainingSet
    
def generateNoise(tSet):
    lenTSet = len(tSet)
    tenPercentSize = int(lenTSet*0.1)
    for i in range(tenPercentSize):
        n = np.random.randint(lenTSet)
        tSet['f'][n] = tSet['f'][n] * (-1)
        
```

#### Ploting the training set

```python
N = 1000
f, trainingSet = createTrainingSet(N)
generateNoise(trainingSet)


colors = np.where(trainingSet['f'] == -1, 'r', 'k' )
plt.scatter(trainingSet['x1'], trainingSet['x2'], color= colors)
plt.show()
```

### Exercice 8

#### Solving Linear Regresion

```python
def solveLinearRegresion(X, y):
    #X = np.array(trainingSet.reindex(columns = cols))
    X_pseudo_inverse = inv(X.T.dot(X)).dot(X.T)
    y_hat = X_pseudo_inverse.dot(y)
    return y_hat

def getVector(x1, x2):
    l = x1.size
    if l == 1:
        return np.array([1, x1, x2])
    else:
        return np.array([np.ones(l), x1, x2])

def getValue(w, x1, x2):
    p1 = getVector(x1, x2)
    return np.sign(w.dot(p1))

def evaluateEin(trainingSet, y_hat, N):
    g_trainingSet = np.array([getValue(y_hat, x1, x2) for [x0, x1, x2, _] in trainingSet.values])
    f_trainingSet = np.array(trainingSet['f'])
    return 1.0*sum(f_trainingSet != g_trainingSet)/N
```

```python
def experiment8():
    N = 1000
    f, trainingSet = createTrainingSet(N)
    generateNoise(trainingSet)
    #Solving linear regresion
    tSetFeatures = trainingSet.loc[:,['x0','x1','x2']].values
    y_hat = solveLinearRegresion(tSetFeatures, trainingSet['f'])
    return evaluateEin(trainingSet, y_hat, N)
```

```python
np.mean([experiment8() for i in range(1000)])
```

### Exercice 9

#### transform the N = 1000 training data into the following nonlinear feature vector: (1, x1, x2, x1x2, x1^2, x2^2)

```python
def createTrainingSet(N):
    trainingSet = pd.DataFrame()
    trainingSet['x0'] = np.ones(N)
    trainingSet['x1'] = np.random.uniform(low=-1, high=1, size=N )
    trainingSet['x2'] = np.random.uniform(low=-1, high=1, size=N )
    trainingSet['x1x2'] = trainingSet['x1']*trainingSet['x2']
    trainingSet['x1^2'] = trainingSet['x1']*trainingSet['x1']
    trainingSet['x2^2'] = trainingSet['x2']*trainingSet['x2']
    f = createTargetFuntion()
    trainingSet['f'] = [f(x1,x2) for [x0, x1, x2, x1x2, x1x1, x2x2] in trainingSet.values]
    return f,trainingSet
```

```python
def experiment9():
    N = 1000
    f,tSet = createTrainingSet(N)
    generateNoise(tSet)
    tSetFeatures = tSet.loc[:,['x0','x1','x2','x1x2','x1^2','x2^2']].values
    return solveLinearRegresion(tSetFeatures,tSet['f'])
```

```python
np.mean([experiment9() for i in range(50)], axis=0)
```

```python
def createTestSet(N):
    testSet = pd.DataFrame()
    testSet['x0'] = np.ones(N)
    testSet['x1'] = np.random.uniform(low=-1, high=1, size=N )
    testSet['x2'] = np.random.uniform(low=-1, high=1, size=N )
    # adding non linear features
    testSet['x1x2'] = testSet['x1']*testSet['x2']
    testSet['x1^2'] = testSet['x1']*testSet['x1']
    testSet['x2^2'] = testSet['x2']*testSet['x2']
    f = createTargetFuntion()
    testSet['f'] = [f(x1,x2) for [x0, x1, x2, x1x2, x1x1, x2x2] in testSet.values]
    return testSet

def getValue(w, vector):
    p1 = np.array(vector)
    return np.sign(w.dot(p1))

#def evaluateEout(w, testSet, y_hat, N):
#    f_testSet = np.array([getValue(w, x0, x1, x2) for [x0, x1, x2, x1x2, x1x1, x2x2] in testSet.values])
#    g_testSet = np.array([getValue(y_hat, x1, x2) for [x0, x1, x2, _] in testSet.values])
#    return 1.0*sum(f_testSet != g_testSet)/N
```

```python
def experiment10():
    N1 = 1000
    NTest = 1000
    f,tSet1 = createTrainingSet(N1)
    generateNoise(tSet1)
    tSetFeatures = tSet1.loc[:,['x0','x1','x2','x1x2','x1^2','x2^2']].values
    y_hat = solveLinearRegresion(tSetFeatures, tSet1['f'])
    testSet = createTestSet(NTest)
    generateNoise(testSet)
    testSetFeatures = testSet.loc[:,['x0','x1','x2','x1x2','x1^2','x2^2']].values
    g_testSet = np.sign(testSetFeatures.dot(y_hat))
    E_out = 1.0*sum(testSet['f'] != g_testSet) / NTest
    return E_out
```
